```python
import osmnx as ox
import networkx as nx
import dimod
from openjij import SASampler
import matplotlib.pyplot as plt
import random

# グラフを取得（例: 宇治市）
place_name = "Uji, Kyoto, Japan"
graph = ox.graph_from_place(place_name, network_type="walk")

# ノードに標高情報を割り当てる（ランダムに設定）
for node, data in graph.nodes(data=True):
    data['elevation'] = random.uniform(0, 100)  # 標高を0～100mの範囲でランダムに設定

# スタートとゴール地点の座標を指定（例: 経度135.76, 緯度34.89）
start = (135.7980142, 34.8786947)  # 経度、緯度を手動で入力
goal = start  # スタートとゴールを同じ座標に設定 34.8884632, 135.7635596

import itertools

for order in itertools.permutations(range(3), 2):
    if isinstance(selected_nodes[0], dict):
       pass_through_1 = (selected_nodes[order[0]]["longitude"], selected_nodes[order[0]]["latitude"])
       pass_through_2 = (selected_nodes[order[1]]["longitude"], selected_nodes[order[1]]["latitude"])

    elif isinstance(selected_nodes[0], tuple):
       pass_through_1 = (selected_nodes[order[0]][2], selected_nodes[order[0]][1])
       pass_through_2 = (selected_nodes[order[1]][2], selected_nodes[order[1]][1])
    else:
         raise TypeError("選択されたノードの形式が不正です。")

# スタート地点、通過地点、ゴール地点のノードを取得
start_node = ox.distance.nearest_nodes(graph, start[0], start[1])  # スタート
goal_node = ox.distance.nearest_nodes(graph, goal[0], goal[1])  # ゴール
pass_through_node_1 = ox.distance.nearest_nodes(graph, pass_through_1[0], pass_through_1[1])  # 通過点1
pass_through_node_2 = ox.distance.nearest_nodes(graph, pass_through_2[0], pass_through_2[1])  # 通過点2

def build_qubo(graph, start_node, goal_node):
    qubo = {}
    elevation_weight = 800 #標高（elevation）に関する重み係数
    penalty_start_goal = 1000 #スタート地点・ゴール地点に関する制約違反のペナルティ
    penalty_continuity = 100 #経路の連続性に関するペナルティ
    distance_weight = 100  #距離に関する重み係数

  # 標高値の正規化
    elevations = [graph.nodes[node]['elevation'] for node in graph.nodes()]
    min_elevation = min(elevations)
    max_elevation = max(elevations)

    def normalize(elevation):
        if max_elevation == min_elevation:
            return 0.0
        return (elevation - min_elevation) / (max_elevation - min_elevation)

    # 目的関数（標高の低いルートを選択する）
    for u, v, data in graph.edges(data=True):
        elevation_u = normalize(graph.nodes[u]['elevation'])
        elevation_v = normalize(graph.nodes[v]['elevation'])
        qubo[(u, v)] = qubo.get((u, v), 0) + elevation_weight * (elevation_u + elevation_v) / 2

    # 制約: スタートとゴールを含むノード
    qubo[(start_node, start_node)] = qubo.get((start_node, start_node), 0) - penalty_start_goal
    qubo[(goal_node, goal_node)] = qubo.get((goal_node, goal_node), 0) - penalty_start_goal

    # 制約: 連続性の保証
    for u, v, data in graph.edges(data=True):
        qubo[(u, v)] = qubo.get((u, v), 0) - penalty_continuity

    # 距離にも重みをつける場合
    for u, v, data in graph.edges(data=True):
        length = data.get('length', 1)
        qubo[(u, v)] = qubo.get((u, v), 0) - distance_weight * length

    return qubo


# QUBOを解く関数
def solve_qubo(graph, qubo, start_node, goal_node):
    sampler = SASampler()
    response = sampler.sample_qubo(qubo)

    # 選択されたノードを取得
    best_sample = response.first.sample
    selected_nodes = [node for node, value in best_sample.items() if value == 1]

    # 選択されたノード間でルートを再構築
    subgraph = graph.subgraph(selected_nodes)
    try:
        route = nx.shortest_path(subgraph, source=start_node, target=goal_node)
        return route
    except nx.NetworkXNoPath:
        print("選択されたノード間に有効な経路が存在しません。")
        return []


import itertools

# selected_nodes: 3点のリスト（dict or tuple形式どちらにも対応）
perms = []

# 3点から2点を選ぶ順列（順序あり全探索）
for a, b in itertools.permutations(range(3), 2):
    perms.append((a, b))

routes = []

for order in perms:
    idx_a, idx_b = order
    if isinstance(selected_nodes[0], dict):
        pass_through_a = (selected_nodes[idx_a]["longitude"], selected_nodes[idx_a]["latitude"])
        pass_through_b = (selected_nodes[idx_b]["longitude"], selected_nodes[idx_b]["latitude"])
    elif isinstance(selected_nodes[0], tuple):
        pass_through_a = (selected_nodes[idx_a][2], selected_nodes[idx_a][1])
        pass_through_b = (selected_nodes[idx_b][2], selected_nodes[idx_b][1])
    else:
        raise TypeError("選択されたノードの形式が不正です。")

    # ノードID取得
    pass_through_node_a = ox.distance.nearest_nodes(graph, pass_through_a[0], pass_through_a[1])
    pass_through_node_b = ox.distance.nearest_nodes(graph, pass_through_b[0], pass_through_b[1])

    # スタート→A
    qubo_1 = build_qubo(graph, start_node, pass_through_node_a)
    route_1 = solve_qubo(graph, qubo_1, start_node, pass_through_node_a)
    # A→B
    qubo_2 = build_qubo(graph, pass_through_node_a, pass_through_node_b)
    route_2 = solve_qubo(graph, qubo_2, pass_through_node_a, pass_through_node_b)
    # B→ゴール（スタートと同じ）
    qubo_3 = build_qubo(graph, pass_through_node_b, goal_node)
    route_3 = solve_qubo(graph, qubo_3, pass_through_node_b, goal_node)

    # ルートを結合
    full_route = route_1 + route_2[1:] + route_3[1:]

    # ルートの総距離を計算
    route_length = 0
    for i in range(len(full_route) - 1):
        edge_data = graph.get_edge_data(full_route[i], full_route[i + 1], default={})
        if edge_data:
            for key, value in edge_data.items():
                if 'length' in value:
                    route_length += value['length']

    routes.append({
        "order": order,
        "full_route": full_route,
        "route_length": route_length,
        "nodes": (pass_through_node_a, pass_through_node_b)
    })

# 目標距離に一番近いルートを選択
target_distance = 9500
best_route = min(routes, key=lambda x: abs(x["route_length"] - target_distance))

print(f"最適な組み合わせ: 通過点{best_route['order'][0]+1}→通過点{best_route['order'][1]+1}")
print(f"ルート距離: {best_route['route_length']:.2f} m")

# 距離条件を満たしているかチェック
target_distance = 9500  # 目標距離（単位: メートル）
tolerance = 0.2 * target_distance  # 許容誤差
lower_bound = target_distance - tolerance
upper_bound = target_distance + tolerance

if lower_bound <= route_length <= upper_bound:
    print(f"ルートが条件を満たしました。距離: {route_length:.2f} m")
else:
    print(f"ルートが条件を満たしません。距離: {route_length:.2f} m")
    print(f"目標距離の範囲: {lower_bound:.2f} m ～ {upper_bound:.2f} m")

# 地図上にルートをプロット
fig, ax = ox.plot_graph_route(graph, full_route, route_linewidth=2, node_size=0, bgcolor='white')
plt.show()
```

```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np

# 1. グラフ取得
place_name = "Uji, Kyoto, Japan"
graph = ox.graph_from_place(place_name, network_type="walk")
for node, data in graph.nodes(data=True):
    data['elevation'] = random.uniform(0, 100)

# 2. 動物の顔パターンを作成（例：猫の顔-輪郭＋目＋鼻＋口）
center_lon, center_lat = 135.79, 34.88  # 宇治市内の中心（例）
face_radius = 0.006  # 輪郭の大きさ（度単位）
num_outline = 50     # 輪郭の点数

# 輪郭（円）
angles = np.linspace(0, 2*np.pi, num_outline, endpoint=False)
outline_points = [
    (center_lon + face_radius * np.cos(a), center_lat + face_radius * np.sin(a))
    for a in angles
]

# 目・鼻・口
eye_offset_x, eye_offset_y = 0.002, 0.002
nose_offset_y = 0.001
mouth_offset_y = -0.002

eye_left = (center_lon - eye_offset_x, center_lat + eye_offset_y)
eye_right = (center_lon + eye_offset_x, center_lat + eye_offset_y)
nose = (center_lon, center_lat + nose_offset_y)
mouth = (center_lon, center_lat + mouth_offset_y)

pattern_points = outline_points + [eye_left, eye_right, nose, mouth]

# 3. 各パターン点に最も近いノードを取得
pattern_nodes = []
for lon, lat in pattern_points:
    try:
        node = ox.distance.nearest_nodes(graph, lon, lat)
        pattern_nodes.append(node)
    except Exception:
        continue  # 点が見つからない場合はスキップ

# 4. ルート生成（パターン点を順に通る経路）
full_route = []
total_length = 0
for i in range(len(pattern_nodes)-1):
    # 各区間で最短経路を取得
    try:
        route = nx.shortest_path(graph, pattern_nodes[i], pattern_nodes[i+1], weight="length")
        if i > 0:
            route = route[1:]  # 2回目以降は重複ノードカット
        full_route.extend(route)
        # 距離計算
        total_length += sum(
            graph[u][v][0].get('length',1) for u,v in zip(route[:-1], route[1:])
        )
    except Exception:
        continue  # 経路が取れない区間はスキップ

# 5. スタートとゴールを同じにしたい場合は、最後にスタートへ戻る経路も加える
try:
    route = nx.shortest_path(graph, pattern_nodes[-1], pattern_nodes[0], weight="length")
    full_route.extend(route[1:])
    total_length += sum(
        graph[u][v][0].get('length',1) for u,v in zip(route[:-1], route[1:])
    )
except Exception:
    pass

# 6. 制約条件のチェック（距離、標高など）
target_distance = 10000
tolerance = 0.2 * target_distance
if abs(total_length - target_distance) < tolerance:
    print(f"距離条件OK: {total_length:.2f} m")
else:
    print(f"距離条件NG: {total_length:.2f} m (目標: {target_distance}±{tolerance})")

# 7. 地図上にコースを描画
fig, ax = ox.plot_graph_route(graph, full_route, route_linewidth=3, node_size=0, bgcolor='white')
plt.show()
```
